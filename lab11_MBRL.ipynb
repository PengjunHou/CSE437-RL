# Gaussian Process

```python
import torch
import gpytorch
import numpy as np

# ============================================================
# ---- Base Exact GP model -------------------
# ============================================================

class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood, kernel='RBF', ard_dims=None):
        super().__init__(train_x, train_y, likelihood)
        self.mean_module = gpytorch.means.ConstantMean()

        # Choose kernel type
        if kernel == 'RBF':
            base_kernel = gpytorch.kernels.RBFKernel(ard_num_dims=ard_dims)
        elif kernel == 'Matern':
            base_kernel = gpytorch.kernels.MaternKernel(nu=2.5, ard_num_dims=ard_dims)
        elif kernel == 'RQ':
            base_kernel = gpytorch.kernels.RationalQuadraticKernel(ard_num_dims=ard_dims)
        else:
            raise ValueError(f"Unsupported kernel type: {kernel}")

        self.covar_module = gpytorch.kernels.ScaleKernel(base_kernel)

    def forward(self, x):
        mean_x = self.mean_module(x)
        covar_x = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean_x, covar_x)


# ============================================================
# ---- Individual GP Manager: handles data + normalization ----
# ============================================================

class GPManager:
    def __init__(self, kernel='RBF', lr=0.03, iters=1000):
        self.kernel = kernel
        self.lr = lr
        self.iters = iters
        self.trained = False
        self.X_train = []
        self.Y_train = []

        self.likelihood = None
        self.model = None


    # ----------------------------- #
    #        FIT / INITIAL TRAIN    #
    # ----------------------------- #
    def fit(self, X, Y):
        X = torch.tensor(X, dtype=torch.float32)
        Y = torch.tensor(Y, dtype=torch.float32)

        self.X_train = X.clone()
        self.Y_train = Y.clone()

        self.retrain()

    def retrain(self):
        # Normalize the input/output and then train the model
        self._compute_normalization()
        self._train_model()

    # ----------------------------- #
    #       ADD NEW DATA POINTS     #
    # ----------------------------- #
    def add_data(self, X_new, Y_new, retrain=True):
        """
        Add new data (single or batch) to GP training set.
        Optionally retrain the model.
        """
        X_new = torch.tensor(X_new, dtype=torch.float32)
        Y_new = torch.tensor(Y_new, dtype=torch.float32).flatten()  # ensure 1D

        # Also flatten stored Y to ensure shape consistency
        if self.Y_train.ndim > 1:
            self.Y_train = self.Y_train.flatten()

        # Append new data
        self.X_train = torch.cat([self.X_train, X_new], dim=0)
        self.Y_train = torch.cat([self.Y_train, Y_new], dim=0)

        if retrain:
            self.retrain()

    # ----------------------------- #
    #         INTERNAL UTILS        #
    # ----------------------------- #
    def _compute_normalization(self):
        self.X_mean, self.X_std = self.X_train.mean(0), self.X_train.std(0)
        self.Y_mean, self.Y_std = float(self.Y_train.mean(0)), float(self.Y_train.std(0))

        Xn = (self.X_train - self.X_mean) / self.X_std
        Yn = (self.Y_train - self.Y_mean) / self.Y_std

        self.Xn, self.Yn = Xn, Yn

    def dataset(self):
        # Extract training data (already stored inside the wrapper)
        X_train = self.X_train.numpy()
        Y_train = self.Y_train.numpy()
        return X_train, Y_train


    def _train_model(self):
        self.likelihood = gpytorch.likelihoods.GaussianLikelihood()
        self.likelihood.noise_covar.initialize(noise=1e-3)
        self.model = ExactGPModel(self.Xn, self.Yn, self.likelihood,
                                  kernel=self.kernel, ard_dims=self.X_train.shape[-1])

        # -------------------------------------------------
        # Train the model
        # -------------------------------------------------
        self.train_gp(self.model, self.likelihood, self.Xn, self.Yn)
        self.trained = True

    def train_gp(self, model, likelihood, x, y):
        model.train(); likelihood.train()
        opt = torch.optim.Adam(model.parameters(), lr=self.lr)
        mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)
        for _ in range(self.iters):
            opt.zero_grad()
            out = model(x)
            loss = -mll(out, y)
            loss.backward()
            opt.step()
        model.eval(); likelihood.eval()

    def predict(self, X):
        if not self.trained:
            raise RuntimeError("GP has not been trained yet.")

        # Normalize the input
        X = torch.as_tensor(X, dtype=torch.float32)
        Xn = (X - self.X_mean) / self.X_std
        with torch.no_grad(), gpytorch.settings.fast_pred_var():
            pred = self.likelihood(self.model(Xn))

            # Unonormalize the output
            mean = pred.mean.cpu().numpy() * self.Y_std + self.Y_mean
            var = pred.variance.cpu().numpy() * (self.Y_std ** 2)
        # Return the unnormalized output
        return mean, var
```

# Setup environment and collect initial data

```python
import gymnasium as gym
# import gym
import numpy as np
import torch
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1.  Environment setup
# --------------------------------------------------------------
env = gym.make("MountainCarContinuous-v0", max_episode_steps=500)
n_state = env.observation_space.shape[0]    # [position, velocity]
n_action = env.action_space.shape[0]        # 1 continuous action
n_output = 2                                # Δpos, Δvel


# --------------------------------------------------------------
# 2.  Collect initial random data
# --------------------------------------------------------------
data = {'X': [], 'Y': []}

obs, _ = env.reset(seed=0)
for _ in range(200):  # collect 500 transitions
    a = env.action_space.sample()  # random continuous action in [-1,1]
    x_next, r, term, trunc, _ = env.step(a)
    data['X'].append(np.concatenate([obs, a]))   # [pos, vel, action]
    data['Y'].append(x_next - obs)               # Δstate
    obs = x_next
    if term or trunc:
        obs, _ = env.reset()

X = np.array(data['X'])[::3]
Y = np.array(data['Y'])[::3]
print(f"Collected {len(X)} transitions.")


# --------------------------------------------------------------
# 3.  Train GP dynamics models
# --------------------------------------------------------------
gps = [GPManager(kernel='RBF', iters=300) for _ in range(n_output)]

for d in range(n_output):
    gps[d].fit(X, Y[:, d])
    print(f"Trained GP for Δstate[{d}] with {len(X)} samples.")
```

# MPC


1. Model-Based RL learns an explicit dynamics model and can use it to simulate many virtual rollouts for planning, so it can evaluate and improve policies with far fewer real environment interactions than Q-learning, which relies purely on sampled transitions. Additionally, by planning over predicted future states, MBRL can “reuse” each collected transition many times in the planner, leading to much higher sample efficiency than model-free methods like Q-learning.

2. The downside of MBRL is model bias: if the learned dynamics model is inaccurate, the planner will optimize with respect to the wrong dynamics and may produce poor actions in the real environment.

```python
#  GP-based dynamics model: s_{t+1} = s_t + Δs
GOAL_POS = 0.45          # MountainCarContinuous goal
POS_MIN, POS_MAX = -1.2, 0.6
VEL_MIN, VEL_MAX = -0.07, 0.07

def gp_step(state, action, gps):
    """
    predict with trained model
    """
    pos, vel = state
    x_in = np.array([[pos, vel, action]], dtype=np.float32)

    delta = np.zeros(2, dtype=np.float32)
    for d in range(2):
        mean, var = gps[d].predict(x_in)
        delta[d] = mean[0]

    next_state = state + delta

    # keep valid
    next_pos = np.clip(next_state[0], POS_MIN, POS_MAX)
    next_vel = np.clip(next_state[1], VEL_MIN, VEL_MAX)
    return np.array([next_pos, next_vel], dtype=np.float32)


#  Cost function for MPC
def stage_cost(state, action):
    pos, vel = state
    # err to goal pos
    pos_cost = (GOAL_POS - pos) ** 2
    vel_cost = 0.1 * vel ** 2
    act_cost = 0.001 * action ** 2
    return pos_cost + vel_cost + act_cost


def terminal_cost(state):
    pos, _ = state
    if pos >= GOAL_POS:
        return 0.0
    else:
        return 10.0 * (GOAL_POS - pos) ** 2

#  MPPI-Simplified / Random Shooting MPC controller
def mpc_action(state,
               gps,
               horizon=25,
               num_samples=800,
               action_low=-1.0,
               action_high=1.0):

    state = np.array(state, dtype=np.float32)

    action_seqs = np.random.uniform(
        low=action_low,
        high=action_high,
        size=(num_samples, horizon)
    )

    costs = np.zeros(num_samples, dtype=np.float32)

    for k in range(num_samples):
        s = state.copy()
        total_cost = 0.0

        for t in range(horizon):
            a = float(action_seqs[k, t])

            # stage cost
            total_cost += stage_cost(s, a)

            # next state
            s = gp_step(s, a, gps)

        # terminal cost
        total_cost += terminal_cost(s)
        costs[k] = total_cost

    best_idx = np.argmin(costs)
    best_first_action = float(action_seqs[best_idx, 0])

    best_first_action = np.clip(best_first_action, action_low, action_high)

    return best_first_action
```

```python
#  GP dynamics (mean + variance)

def gp_step_with_var(state, action, gps):
    """
    next_state_mean: shape (2,)
    next_state_var:  shape (2,)  —— GP predictive variance of Δs
    """
    pos, vel = state
    x_in = np.array([[pos, vel, action]], dtype=np.float32)

    delta_mean = np.zeros(2, dtype=np.float32)
    delta_var = np.zeros(2, dtype=np.float32)

    for d in range(2):
        mean, var = gps[d].predict(x_in)
        delta_mean[d] = mean[0]
        delta_var[d] = var[0]     # variance of Δs

    next_state_mean = state + delta_mean
    next_state_var = delta_var   # not added into state, but used for cost

    next_state_mean[0] = np.clip(next_state_mean[0], POS_MIN, POS_MAX)
    next_state_mean[1] = np.clip(next_state_mean[1], VEL_MIN, VEL_MAX)

    return next_state_mean, next_state_var

#  Stage cost with uncertainty penalty
UNC_PENALTY = 0.5  # how strongly we penalize uncertainty

def stage_cost_uncertainty(state, action, delta_var):
    """
    state      = [pos, vel]
    delta_var  = variance of GP prediction for Δs (shape (2,))
    """
    base_cost = stage_cost(state, action)
    unc_cost = UNC_PENALTY * np.sum(delta_var)  # penalize
    return base_cost + unc_cost


#  MPPI with GP uncertainty (full Exercise 2 version)
def mpc_action_uncertain(state,gps,horizon=25,num_samples=600,action_low=-1.0,action_high=1.0):
    """
    add variance penalty。
    """

    state = np.array(state, dtype=np.float32)
    action_seqs = np.random.uniform(
        low=action_low,
        high=action_high,
        size=(num_samples, horizon)
    )

    costs = np.zeros(num_samples, dtype=np.float32)

    for k in range(num_samples):
        s = state.copy()
        total_cost = 0.0

        for t in range(horizon):
            a = float(action_seqs[k, t])

            # --- Use GP dynamics + predictive variance ---
            next_mean, next_var = gp_step_with_var(s, a, gps)

            # --- cost including variance penalty ---
            total_cost += stage_cost_uncertainty(s, a, next_var)

            # forward GP mean
            s = next_mean

        total_cost += terminal_cost(s)
        costs[k] = total_cost

    best = np.argmin(costs)
    a0 = float(action_seqs[best, 0])
    return np.clip(a0, action_low, action_high)
```

```python

#  Run both controllers and compare

def run_episode(env, gps, use_uncertainty=False, max_steps=500):
    obs, _ = env.reset()
    positions = [obs[0]]

    for t in range(max_steps):
        if use_uncertainty:
            a = mpc_action_uncertain(obs, gps)
        else:
            a = mpc_action(obs, gps)

        next_obs, reward, terminated, truncated, _ = env.step(np.array([a]))
        positions.append(next_obs[0])
        obs = next_obs

        if terminated or truncated:
            break

    return positions, t+1  # return number of steps

```
